```python
import re
```

```python
import pandas as pd
```

```python
messy_of_words = pd.DataFrame(pd.read_csv("other_disease_list.csv"))
messy_of_words
```

```python
messy_of_words["MORB_DESC"].apply(lambda x: x.split(','))
```

```python
word_set = set()
for i in messy_of_words["MORB_DESC"].apply(lambda x: re.split(r'[,|;|,|/|+|]'+'| E ',x)):
    word_set.update(i)
```

```python
wordfreq = dict.fromkeys(word_set,0)
for i in messy_of_words["MORB_DESC"].apply(lambda x: re.split(r'[,|;|,|/|+|]'+'| E ',x)):
    for j in i:
        if j in wordfreq.keys():
            wordfreq[j] += 1
```

```python
fr = pd.Series(wordfreq).sort_values(ascending=False)
```
