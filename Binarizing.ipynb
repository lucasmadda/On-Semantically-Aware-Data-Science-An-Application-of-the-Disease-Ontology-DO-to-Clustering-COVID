```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Data

```python
labels = pd.read_csv("labels.csv",index_col=[0]).iloc[:,0]
data = pd.read_csv("data.csv",index_col = [0])
words = pd.read_csv("rulefiles/words.csv",index_col = [0]).iloc[:,0].apply(lambda x: x[1:-1].replace("'","").replace("","").split(","))
columns = pd.read_csv("rulefiles/columns.csv",index_col = [0]).iloc[:,0]
ancestors = pd.read_csv("rulefiles/ancestors.csv",index_col = [0]).iloc[:,0].apply(lambda x: x[1:-1].replace("'","").replace("","").split(","))
data["MORB_DESC"].fillna("", inplace = True)
data["MORB_DESC"] = data["MORB_DESC"].apply(lambda x: re.split(r'[,|;|,|/|+|]'+'| E ',x))
```

```python
word_set = set()
for i in data["MORB_DESC"]:
    word_set.update(i)
```

# Creating binary dataset - no ancestrality respected

```python
bin_data = pd.DataFrame(index = data.index,
                       columns = words.keys(),
                       data = 0)
for col in columns.index:
    col_name = columns[col]
    bin_data[col] = data[col_name]
    
for col in words.index:
    for word in words[col]:
        bin_data[col] += data['MORB_DESC'].apply(lambda x: 1 if word in x else 0)
    bin_data[col] = bin_data[col].apply(lambda x: 1 if x!=0 else 0)

bin_data
```

# Creating binary dataset - ancestrality respected

```python
words_a = words
for descendant in ancestors.keys():
    for ancestor in ancestors[descendant]:
        if type(ancestor) == list:
            print(ancestor)
            words_a[ancestor] += words.to_dict().get(descendant,[])

bin_data_a = pd.DataFrame(index = data.index,
                       columns = words_a.keys(),
                       data = 0)
for col in columns.index:
    col_name = columns[col]
    bin_data_a[col] = data[col_name]
    
for col in words_a.index:
    for word in words_a[col]:
        bin_data_a[col] += data['MORB_DESC'].apply(lambda x: 1 if word in x else 0)
    bin_data_a[col] = bin_data_a[col].apply(lambda x: 1 if x!=0 else 0)
    
bin_data_a
```

# Analyzing difference when considering ancestrality

```python
diffs = (bin_data_a - bin_data).sum()
diffs[diffs != 0].rename(labels)
```

# Exporting FIles

```python
bin_data.to_csv("binary_no_ancestrality.csv")
bin_data_a.to_csv("binary_with_ancestrality.csv")
```
