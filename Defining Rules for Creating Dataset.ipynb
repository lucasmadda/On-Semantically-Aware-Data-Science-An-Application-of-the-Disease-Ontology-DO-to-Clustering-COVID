# Get all ancestors of a given class

```python
def is_a_list(c):
    return [i for i in c.is_a if str(i)[:8] == 'obo.DOID']

def all_ancestors(c):
    ancestors = is_a_list(c)
    for i in ancestors:
        if i == owl.Thing or i ==[]:
            pass
        else:
            ancestors += is_a_list(i)        
    return ancestors

def all_ancestors_in_context(c,context):
    return [i for i in all_ancestors(c) if i in context]
```
